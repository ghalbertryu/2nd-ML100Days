1. 線性回歸模型能夠準確預測非線性關係的資料集嗎?
2. 回歸模型是否對資料分布有基本假設?


### Ans:
#### 1. 非線性關係的問題使用線性回歸大部分情況是無法準確預測的，因為線性回歸將模型過度簡化，遇到非線性的數據就無法完整去 fit 數據。使用梯度下降更能準確預測線性及非線性資料。
#### 2. 有基礎假設；Linear Regression 是假設資料呈線性關係，Logistics Regression 是假設資料呈現二元分類型態。
