```python
import os
import numpy as np
import pandas as pd

dir_data = '../data/data-science-london-scikit-learn/'
```

```python
# Read train X
train_X_data = os.path.join(dir_data, 'train.csv')
train_X = pd.read_csv(train_X_data, header=None)
train_X.head()
```

```python
train_X.describe()
```

```python
# Read train Y
train_Y_data = os.path.join(dir_data, 'trainLabels.csv')
train_Y = pd.read_csv(train_Y_data, header=None)
train_Y.head()
```

```python
# Read test X
test_X_data = os.path.join(dir_data, 'test.csv')
test_X = pd.read_csv(test_X_data, header=None)
test_X.head()
```

```python
test_X.describe()
```

```python
# Split train set and test set
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_X, train_Y, test_size=0.1, random_state=42)
x_train.describe()
```

```python
# train Model
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
```

```python
clf = GradientBoostingClassifier(n_estimators=900, 
                                 max_depth=7)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
# predict test set by Model
pred_test_Y = clf.predict(test_X)
pred_test_Y
```

```python
# format predict data
pred_df = pd.DataFrame(pred_test_Y)
pred_df.index.name = 'Id'
pred_df.columns = ['Solution']
pred_df = pred_df.reset_index()
pred_df['Id'] = pred_df['Id'] + 1
pred_df.head()
```

```python
pred_df.to_csv("P01_london-scikit-learn.csv", index=False)
```
